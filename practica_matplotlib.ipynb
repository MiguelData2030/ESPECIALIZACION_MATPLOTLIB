# CLASE 2  graficos de barras .bar

```python

df=pd.read_csv('data.csv') # leemos el archivo

language_counter= Counter()  #aplicamos funcion Couter]

for x in df['LanguagesWorkedWith']:               #Realizamos un for que itera por la columna 
     language_counter.update(x.split(';'))    

languages=[]
users=[]

for item in language_counter.most_common(15):      #Realizamos otro for para sacar el total x lenguaje y Usuarios
     languages.append(item[0])
     users.append(item[1])

languages.reverse()                               #Esta funcion nos ayuda a organizar las barras de mayor a menor                    
users.reverse()
plt.barh(languages,users)
plt.title('Lenguajes mas populares')              # barth me permite invertir las barras
plt.xlabel('Cantidad Usuarios ', fontsize=14)
#plt.ylabel('Lenguajes de Programacion')

plt.tight_layout()
plt.show()
```

```python


plt.style.use('seaborn')

data = pd.read_csv('2019-05-31-data.csv')   #traemos el archivo csv
view_count = data['view_count']             # declaramos variables y nos traemos las columnas dle dataframe
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count,likes, c=ratio,linewidths=1, cmap='inferno',      # C= ratio color al ratio, cmap es la paleta que le queremos dar al grafico
            edgecolors='black',alpha=0.90)     #para este eje,plo escojemos una paleta que muestre bien los degradados ya que es un grafico de dispersion
cbar=plt.colorbar()                    #Activa la barra lateral que es el termometro dle lado derecho del scatter                         
cbar.set_label('Like/Dislike Ratio')   #Titulo para el termometro


plt.xscale('log')     #Se aplica escala logaritmica para visualizar mejor el impacto ya que hay unos pocos datos que estan aislados y desconfiguran el grafico                                              
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.tight_layout()
plt.show()
```

# CLASE 9 PLOTTING LIVE DATA IN REAL TIME

```python

import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('fivethirtyeight')
```

```python
import csv
import random
import time

x_value = 0
total_1 = 1000
total_2 = 1000

fieldnames = ["x_value", "total_1", "total_2"]


with open('data.csv', 'w') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()

while True:

    with open('data.csv', 'a') as csv_file:
        csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

        info = {
            "x_value": x_value,
            "total_1": total_1,
            "total_2": total_2
        }

        csv_writer.writerow(info)
        print(x_value, total_1, total_2)

        x_value += 1
        total_1 = total_1 + random.randint(-6, 8)
        total_2 = total_2 + random.randint(-5, 6)

    time.sleep(1)
```

```python

x_vals = []
y_vals = []

index = count()


def animate(i):
    data = pd.read_csv('data.csv')
    x = data['x_value']
    y1 = data['total_1']
    y2 = data['total_2']

    plt.cla()

    plt.plot(x, y1, label='Channel 1')
    plt.plot(x, y2, label='Channel 2')

    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.tight_layout()
plt.show()
```
